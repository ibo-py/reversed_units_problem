# The correlation betwen numbers and their reversed form

# here we setting the data to use
basicly we are setting a list of nubers in range [0, 100) after that we flipped the digits and got the delta between the numbers wich is in "y" list 

```python


numbers = [i for i in range(0,100)]
stringvalues = [f"{i}" for i in numbers]

x = []
y = []

for i in stringvalues:
    digits_in_strings = list(i)
    digits_in_integer = [int(i) for i in digits_in_strings]
    try:
        delta_digits = digits_in_integer[1] - digits_in_integer[0] # not nessesary
        flipped_number = int(f"{digits_in_integer[1]}{digits_in_integer[0]}")
    except: 
        delta_digits = digits_in_integer[0] - 0
        flipped_number = int(f"{digits_in_integer[0]}0")
    delta_nums = flipped_number - int(i) # main problem's delta
    x.append(delta_digits)
    y.append(delta_nums)

min, max = -20, 80 # ere we setting the max and min values for the cordinates to visualise the data we made
```

# Here we found that the main formula for range [0, 100) is:
y = 9x - 99(x/10 - (x mod 10)/10)


```python
m = (y[5]-y[4])/(x[5]-x[4]) # the sloup
n = -99
Y = lambda x : x*m + n*(((x/10) - (x%10)/10))
```

# Explaining the formula
the formula is verry simple, we just found the sloup of one of the periodic lenear graphs then i found the correlation wich is in every ten numbers the value of the delta add up (-99) so we multyply -99 with the sth that gives us how many ten numbers we have passed wich means (((x/10) - (x%10)/10))
we se how many 10 we have by dividing the number by ten then we get rid of the decimals by (x mod 10)/10
